## Visualisasi Data


Sebagai seorang peneliti, men-visual-kan data merupakan pekerjaan rutin. Namun bagi seorang peneliti pemula, terkadang bingung tentang cara melakukannya. Berikut ini contoh visualisasi data yang diambil dari short course data camp. 

https://www.datacamp.com/community/blog/intermediate-python-for-data-science-matplotlib#gs.XdtUdZI

Visualisasi data merupakan langkah penting yang harus dilalui saat akan melakukan analisis data. Saat dihadapkan pada data-data, makna dibalik data tersebut harus dicari terlebih dahulu, lalu bagikan makna tersebut kepada orang lain melalui visualisasi yang bagus tapi tetap mudah dimengerti.

Python menyediakan berbagai module untuk melakukan visualisasi data. Matplotlib adalah modul bawaannya dengan sub-modul pyplot yang bisa disingkat menjadi plt. 

Contoh kasus, akan dilakukan plot jumlah pertumbuhan penduduk dunia dalam 50 tahun terakhir. Untuk melakukan plot tersebut gunakan plt.plot() dan gunakan daftar jumlah penduduk dan waktu sebagai argumen. Waktu bisa ditempatkan sebagai di sumbu-x dan jumlah penduduk bisa ditempatkan di sumbu-y. Python tidak akan menampilkan plot data tersebut sebelum dipanggil perintah plt.show(). 

```python
import matplotlib.pyplot as plt
year = [1950,1970,1990,2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year,pop)
plt.show()
```

Cara lain untuk menampilkan data adalah dengan scatter plot. Terkadang dengan scatter plot, lebih bisa menampilkan keadaan data yang sesungguhnya dan lebih mudah untuk dipahami. Perintahnya hanya mengganti plt.plot() menjadi plt.scatter().

```python
import matplotlib.pyplot as plt
year = [1950,1970,1990,2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year,pop)
plt.show()
```

## Custom plot

```python
import pandas as pd
import matplotlib.pyplot as plt
## index_col = 0 untuk menghilangkan angka index yang biasa muncul dalam pandas series atau data frame.
df = pd.read_csv('../DATA/matplotlib_scatter-log.csv',index_col =0)

gdp_cap = df['gdp_cap']
gdp_caps = gdp_cap.values
life_exp = df['life_exp']
life_exps = life_exp.values ## mengubah ke format array numpy dari pandas series.
## menambahkan kolom color ke data gapminder.csv tapi tidak di save dalam file-nya.
for lab, row in df.iterrows():
    if row['cont'] == 'Asia':
        df.loc[lab,'color'] = 'red'
    if row['cont'] == 'Europe':
        df.loc[lab,'color'] = 'green'
    if row['cont'] == 'Africa':
        df.loc[lab,'color'] = 'blue'
    if row['cont'] == 'Americas':
        df.loc[lab,'color'] = 'cyan'
    if row['cont'] == 'Oceania':
        df.loc[lab,'color'] = 'black'
col = df['color']
cols = col.values
pop = df['population']/1000000
pops = pop * 2
plt.scatter(x=gdp_caps,y=life_exps,s=pops,c=col,alpha=0.6)
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000],['1k','10k','100k'])
plt.text(1550,71,'India')
plt.text(5700,80,'China')
plt.grid(True)
plt.show()
```
